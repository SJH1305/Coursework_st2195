```python
## Random walk Metropolis algorithm
```

```python
import numpy as np

N = 10000
s = 1
x_0 = 0

# Normal distribution generator:
np.random.seed(12) # reproducibility, numbers don't change 
numbers = np.random.normal(size = 10000, scale = s) 

x = np.zeros(N) # creating array 
x_0 = x[0]

# Computing the ratio:
def log_f(x):
    return np.log(0.5) - np.abs(x) 
    ''' 
    Coverting f(x) into log form based on the 'practical tip' given in the coursework document.
    
    (0.5)exp(-abs(x)) = log(0.5) - abs(x) 
    '''

for i in range (1,N): # initial value = 0
    x_star = np.random.normal(loc = x[i-1], scale = s)
    r = log_f(x_star) - log_f(x[i-1]) # given ratio formula
    u = np.random.uniform(0,1)
    if np.log(u) < r:# resetting u as log 
        x[i] = x_star
    else:
        x[i] = x[i-1]
```

```python
## Constructing histogram with kernel density plot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import numpy as np

# histogram, kde
np.random.seed(12)
numbers = np.random.normal(size = 10000, scale =1) 

sns.histplot(data=numbers, bins='auto',kde = True, stat = 'density', label = 'KDE') # Density ! 
plt.xlabel('X')
plt.ylabel('Density (Y)')
plt.title("Normally distributed values")

# plotting general(exponential) form of f(x) 
x = np.linspace(-4,4,10000)
y = 0.5*np.exp(-np.abs(x))
plt.plot(x,y, color = 'red', label = 'f(x)') 
plt.legend()
plt.savefig("Histogram_Py.png")
```
